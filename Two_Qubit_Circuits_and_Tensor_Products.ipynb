# **Main Code**

```python
# Import libraries
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
# Step 1: Create a quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2) #2 = 0,1

```

```python
# Step 2: Apply gates to demonstrate tensor product states
qc.x(0)  # Apply X gate on qubit 0 → |10>

```

```python
# Step 3: Measure both qubits
qc.measure([0, 1], [0, 1])   #(classic bits, cubic bits)

```

```python
# Step 4: Use AerSimulator (new standard in Qiskit 2.x)
simulator = AerSimulator()

# Transpile the circuit for the simulator
from qiskit import transpile
compiled_circuit = transpile(qc, simulator)

# Run the circuit
job = simulator.run(compiled_circuit, shots=2000)
result = job.result()

# Get counts
counts = result.get_counts()

```

```python
# Step 5: Display circuit and results
print("Measurement Results:", counts)
print(qc.draw('text'))
# Plot histogram (using save and display)
hist = plot_histogram(counts, title="Measurement Histogram")
hist.savefig("histogram_output.png")
from IPython.display import Image
display(Image("histogram_output.png"))
```

**Task 1: Change Gate Combination**
*   Apply a Hadamard gate on both qubits and observe the output probabilities.
*   What pattern do you notice in the histogram?

```python
qc = QuantumCircuit(2,2)

# Apply Hadamard gate on both qubits
qc.h([0, 1])

qc.measure([0, 1], [0, 1])
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=2000)
result = job.result()
counts = result.get_counts()

# Display Results
print("Measurement Results:", counts)
print("\nQuantum Circuit:")
print(qc.draw('text'))

# Plot histogram
hist = plot_histogram(counts, title="Measurement Histogram")
hist.savefig("histogram_output.png")
from IPython.display import Image
display(Image("histogram_output.png"))
```

*The histogram shows approximately equal probabilities for all four outcomes — 00, 01, 10, and 11.
This is because applying Hadamard gates on both qubits creates a superposition of all possible states (|00⟩, |01⟩, |10⟩, |11⟩), each with equal probability (≈ 25%).*

**Task 2: Swap Qubit Roles**
* Apply an X gate on the second qubit instead of the first.
* Compare the output with the original circuit.

```python
qc = QuantumCircuit(2,2)

# Apply X gate on the second qubit instead of the first
qc.x(1)

qc.measure([0, 1], [0, 1])
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=2000)
result = job.result()
counts = result.get_counts()

# Display Results
print("Measurement Results:", counts)
print("\nQuantum Circuit:")
print(qc.draw('text'))

# Plot histogram
hist = plot_histogram(counts, title="Measurement Histogram")
hist.savefig("histogram_output.png")
from IPython.display import Image
display(Image("histogram_output.png"))
```

*In the original circuit, applying X on the first qubit produced the state |10⟩.
Now, applying X on the second qubit produces the state |01⟩.
The histogram shifts accordingly — the bit positions are swapped, showing the effect of qubit order in the circuit.*

**Task 4: Increase Measurement Shots**
* Change the number of shots from 1024 to 4096 in the execute() function.
* Does the distribution become more stable?

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0, 1)

qc.measure([0, 1], [0, 1])
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)

# number of shots from 1024 to 4096
job = simulator.run(compiled_circuit, shots=4096)

result = job.result()
counts = result.get_counts()

# Display Results
print("Measurement Results:", counts)
print("\nQuantum Circuit:")
print(qc.draw('text'))

# Plot histogram
hist = plot_histogram(counts, title="Measurement Histogram")
hist.savefig("histogram_output.png")
from IPython.display import Image
display(Image("histogram_output.png"))
```

*Yes, increasing the number of shots to 4096 makes the measurement distribution more stable and smoother.
This happens because with more repetitions, statistical noise decreases, giving results closer to the theoretical probabilities.*

**Task 5: Add a New Visualization**
* Use plot_bloch_multivector to visualize the combined state of both qubits.
* Observe how the Bloch sphere changes before and after measurement.

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

# Create and prepare the circuit
qc = QuantumCircuit(2, 2)
qc.h([0, 1])  # Apply Hadamard gates
qc.measure([0, 1], [0, 1])

# Get statevector before measurement
qc_state = QuantumCircuit(2)
qc_state.h([0, 1])
state = Statevector.from_instruction(qc_state)

qc.measure([0, 1], [0, 1])
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=2000)
result = job.result()
counts = result.get_counts()

# Display Results
print("Measurement Results:", counts)
print("\nQuantum Circuit:")
print(qc.draw('text'))

# Visualize combined state on Bloch sphere
plot_bloch_multivector(state, title="Bloch Sphere: Before Measurement")
```

*Before measurement, both qubits appear in superposition on the Bloch sphere (typically pointing along the X-axis for Hadamard states).
After measurement, the Bloch vectors collapse to the poles (|0⟩ or |1⟩), indicating definite classical states.*
